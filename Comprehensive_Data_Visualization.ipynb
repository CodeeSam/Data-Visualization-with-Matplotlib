# MATPLOTLIB

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np

x = np.linspace(0, 10, 25)
y = x*x + 2
print(x)
print(y)
```

## To Draw a Line using pylab

```python
pylab.plot(x, y, 'r') #'r' represent red
```

## To Draw Subgraphs using pylab

```python
pylab.subplot(1,2,1) #the values in the parenthesis represent first row, second column and first index
pylab.plot(x,y, 'r--') #this is for the data from where we are plotting. The 3rd parameter represents the color and line style
pylab.subplot(1,2,2)
pylab.plot(y,x, 'g*-')
```

## Drawing a Line using pyplot

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #the values in the bracket controls the left and right, width and height
axes.plot(x,y, 'r')
```

## To Draw Subgraphs using pyplot

```python
fig = plt.figure()

fig, axes = plt.subplots(nrows=1, ncols=2) #the values in the bracket controls the left and right, width and height
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.8]) #you can manipulate the values here to tune the big plot
ax2 = fig.add_axes([0.2, 0.9, 0.3, 0.3]) #you can manipulate the values here to tune the small plot

ax1.plot(x,y, 'r')
ax2.plot(y,x, 'g')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, x**2,'r')
axes.plot(x, x**3,'g')

axes.set_title('MY PLOT')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.legend(['y= x**2', 'y= x**3'], loc=6) #you can manipulate the location of legends by tuning the value of 'loc'
```

```python
fig, axes = plt.subplots(dpi=100)
axes.plot(x, x+1, color='red', alpha=.5) #you can tune the alpha value to manipulate the color thickness of your lines
axes.plot(x, x+2, color='green', linewidth=.5) #using linewidth, you can also manipulate the width of the lines
axes.plot(x, x+3, color='blue')
```

```python
fig, axes = plt.subplots(dpi=100)
axes.plot(x, x+1, color='red', alpha=.5) #you can tune the alpha value to manipulate the thickness of your lines
axes.plot(x, x+2, color='green', lw=.5) #you can also abbreviate the linewidth as lw
axes.plot(x, x+3, color='blue', linestyle=':') #you can also select the style of your line
line, = axes.plot(x, x+5, color='brown', linestyle='-')
line.set_dashes([5, 15, 5, 15]) #we can use the 'set_dashes' attribute to manipulate the line spacing
```

```python
fig, ax = plt.subplots(dpi=100)

ax.plot(x, x+4, color='red', marker='o')
ax.plot(x, x+3, color='blue', marker='*')
ax.plot(x, x+2, color='black', marker='.', markersize=7, markerfacecolor='red') #using markerfacecolor, we can change the color of our marker
ax.plot(x, x+1, color='green', marker='s', markersize=4) #using markersize, we can size the marker
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].plot(x, x**2, x, x**3, lw=2)
ax[0].grid(True)

ax[1].plot(x, x**2, x, x**3, lw=2)
ax[1].set_ylim([0,65]) #you can set the scale of y and x axis using ylim and xlim
ax[1].set_xlim([2,10])
```

## Other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, ax = plt.subplots(1, 4, figsize=(16,5))
ax[0].set_title('SCATTER PLOT')
ax[0].scatter(x, x + 0.25*np.random.randn(len(x)))

ax[1].set_title('STEP PLOT')
ax[1].step(n, n**2, lw=2)

ax[2].set_title('BAR PLOT')
ax[2].bar(n, n**2, alpha=0.8, align='center', width=0.6)
ax[2].set_xlim(0,6)

ax[3].set_title('FILL-BETWEEN PLOT')
ax[3].fill_between(x, x**2, x**3, color='brown', alpha=.8)
```

### Radar Chart

```python
fig = plt.figure(figsize=(7,7))

ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)


t = np.linspace(0, 2*np.pi, 100)
ax.plot(t,t**2, color='red', lw=2)
```

### Histogram

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4))
n = np.random.randn(100000)

ax[0].set_title('HISTOGRAM')
ax[0].hist(n, bins=10) #the bins is how many 'bars' you want

ax[1].set_title('CUMULATIVE DETAILED HISTOGRAM')
ax[1].hist(n, cumulative=True, bins=12)
```

#### n is an array of 100000 randomly generated values. From the above histogram, we can have insight of the values generated. For example, looking critically, we have about 30 000 zeros
#### N.B: Cumulative histogram arrages the trend in ascending order

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mtcars = pd.read_csv('mtcars.csv')
mtcars.head(8)
```

```python
mtcars.info()
```

```python
mtcars.shape
```

### Bar Plot

```python
# we always put the categorical variables on the x axis while the numerical values on the y axis
result = sns.barplot(mtcars.cyl, mtcars.mpg)
plt.show()
```

```python
#we can also plot the above bargraph using another method:
result = sns.barplot(x='cyl', y='mpg', data = mtcars)
plt.show()
```

##### note that the method above gives no warning

```python
# let's do some coloring
result = sns.barplot(x='cyl', y='mpg', data = mtcars, color='blue')
plt.show()
```

```python
# let's use some color palettes
result = sns.barplot(x='am', y='mpg', data = mtcars, palette= 'Set2')
plt.show()
```

```python
# we can use different palette types
result = sns.barplot(x='gear', y='hp', data = mtcars, palette= 'rocket')
plt.show()
```

### Count Plot

```python
# for count plot, we don't need to define y axis
# we are using countplot to count how many of the values(x axis) we have in the dataset

sns.countplot(x='cyl', data=mtcars, palette='Set1')
plt.show()
```

```python
sns.countplot(x='carb', data=mtcars, palette='Set2')
plt.show()
```

```python
# we can also view it horizonatally by defining y axis and leaving out x axis
sns.countplot(y='gear', data=mtcars, palette='rocket')
```

```python
sns.countplot(x='gear', hue='cyl', data=mtcars, palette='Set1')
```

##### In the above, we can deduce that: we have 12 of those vehicles with 3 gears having 8 cylinders and so on... In other words, vehicles with 3 gears and 8 cylinders are 12 in number

### Scatter Plot

```python
#Scatter plot can help us to determine the relationship between two continuous variables
#We'll be using iris dataset which is already inbuilt in seaborn
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
#let's create a scatter plot to show the relationship between petal_length and sepal_length

sns.scatterplot(x='sepal_length', y='petal_length', data=iris)
```

```python
# let's use the hue parameter to classify them

sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species')
```

```python
# we can also classify using the SIZE of the datapoints

sns.scatterplot(x='sepal_length', y='petal_length', data=iris, size='species')
#BIG, BIGGER, BIGGEST
```

```python
# we can also classify using both hue and size
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species', size='species')
```

### Pair Plot

```python
sns.pairplot(iris)
```

```python
#let's make it more fancy
sns.pairplot(iris, hue='species', palette='Set1')
```

#### with the aid of plots like these above, we can easily deduce the best parameter that best categorize the 3  kind of flower based on how separated they are.

### LM Plot

```python
sns.lmplot(x='petal_length', y='petal_width', data=iris)
#The line drawn below is a linear regression line
```

```python
sns.lmplot(x='petal_length', y='petal_width', hue='species', data=iris, markers=['+', '*', '.'])
```

### Box Plot

```python
#we'll be visualizing the species by ttheir sepal width

sns.boxplot(x='species', y='sepal_width', data=iris)
```

#### The lines at the middle of those boxes represent the median, while the upper and lower lines represent the highest and lowest values. The dots represent the outliers(versicolor species don't have outlier value from the above plot)
