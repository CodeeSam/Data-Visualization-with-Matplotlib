```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('MessiRonaldo.csv')
```

```python
df.head()
```

## Trend Analysis

```python
Ronaldo = ['Ronaldo']
Ronaldo_df = df[df.Player.isin(Ronaldo)]
Ronaldo_df
```

```python
Messi = ['Messi']
Messi_df = df[df.Player.isin(Messi)]
Messi_df
```

###### Total La Liga Goals For Ronaldo

```python
Ronaldo_total_Liga_Goals = 0
for goals in Ronaldo_df.Liga_Goals.values.tolist():
    Ronaldo_total_Liga_Goals+=goals
print(Ronaldo_total_Liga_Goals)
```

###### Total La Liga Goals For Messi

```python
Messi_total_Liga_Goals = 0
for goals in Messi_df.Liga_Goals.values.tolist():
    Messi_total_Liga_Goals+= goals
print(Messi_total_Liga_Goals)
```

###### Total UCL Goals For Ronaldo

```python
Ronaldo_total_UCL_Goals = 0
for goals in Ronaldo_df.CL_Goals.values.tolist():
    Ronaldo_total_UCL_Goals+=goals
print(Ronaldo_total_UCL_Goals)
```

###### Total UCL Goals For Messi

```python
Messi_total_UCL_Goals = 0
for goals in Messi_df.CL_Goals.values.tolist():
    Messi_total_UCL_Goals+=goals
print(Messi_total_UCL_Goals)
```

###### Total La Liga Assists For Ronaldo

```python
Ronaldo_total_Liga_Assists = 0
for assists in Ronaldo_df.Liga_Asts.values.tolist():
    Ronaldo_total_Liga_Assists+=assists
print(Ronaldo_total_Liga_Assists)
```

###### Total La Liga Assists For Messi

```python
Messi_total_Liga_Assists = 0
for assists in Messi_df.Liga_Asts.values.tolist():
    Messi_total_Liga_Assists+=assists
print(Messi_total_Liga_Assists)
```

###### Total UCL Assists For Ronaldo

```python
Ronaldo_total_UCL_Assists = 0
for assists in Ronaldo_df.CL_Asts.values.tolist():
    Ronaldo_total_UCL_Assists+=assists
print(Ronaldo_total_UCL_Assists)
```

###### Total UCL Assists For Messi

```python
Messi_total_UCL_Assists = 0
for assists in Messi_df.CL_Asts.values.tolist():
    Messi_total_UCL_Assists+=assists
print(Messi_total_UCL_Assists)
```

### Plot of La Liga Goals and UCL Goals

```python
fig, ax = plt.subplots(1,2)
Players = ['Ronaldo', 'Messi']
La_Liga_pie_values = [Ronaldo_total_Liga_Goals, Messi_total_Liga_Goals]
UCL_pie_values = [Ronaldo_total_UCL_Goals, Messi_total_UCL_Goals]

fig.subplots_adjust(wspace=1, right=2)
ax[0].set_title('La Liga Goals', fontsize=16)
ax[1].set_title('UCL Goals', fontsize=16)
ax[0].pie(La_Liga_pie_values, labels=Players, colors=('brown','red'), startangle=90, autopct='%1.1f%%', shadow=True, radius=1.1, explode=(0,0.05))
ax[1].pie(UCL_pie_values, labels=Players, colors=('brown','red'), startangle=90, autopct='%1.1f%%', shadow=True, radius=1.1, explode=(0,0.05))
```

### Plot of La Liga Assists and UCL Assists

```python
fig, ax = plt.subplots(1,2)
Players = ['Ronaldo', 'Messi']
La_Liga_pie_values1 = [Ronaldo_total_Liga_Assists, Messi_total_Liga_Assists]
UCL_pie_values1 = [Ronaldo_total_UCL_Assists, Messi_total_UCL_Assists]

fig.subplots_adjust(wspace=1, right=2)
ax[0].set_title('La Liga Assists', fontsize=16)
ax[1].set_title('UCL Assists', fontsize=16)
ax[0].pie(La_Liga_pie_values1, labels=Players, colors=('brown','red'), startangle=90, autopct='%1.1f%%', shadow=True, radius=1.1, explode=(0,0.05))
ax[1].pie(UCL_pie_values1, labels=Players, colors=('brown','red'), startangle=90, autopct='%1.1f%%', shadow=True, radius=1.1, explode=(0,0.05))
```
